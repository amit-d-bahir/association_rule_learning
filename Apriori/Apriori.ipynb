```python
# Apriori Algorithm

# Importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Importing the dataset
df = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
```

```python
transactions = []
for i in range(0,7501):
    transactions.append([ str(df.values[i, j]) for j in range(0,20)])
    # We've done this to make this whole list of transaction ready to be the input of our 
    # Apriori function
    # is to set all the products in the transactions as a string
```

```python
transactions
```

```python
# Training Apriori model on our dataset
from apyori import apriori
# our output
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length=2)
# for a product that is purchased 3 times a day
# Thus support = 3*7 / 7500
```

```python
rules
```

```python
# Visualizing the results
results = list(rules)
```

```python
results
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_mod = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df_mod, min_support=0.003, use_colnames=True)

frequent_itemsets
```

```python
# from frequent_patterns.association_rules import association_rules
from mlxtend.frequent_patterns import association_rules
orgrules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
orgrules
```
